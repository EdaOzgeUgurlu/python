```python
🔍 1. Linear Search Egzersizi
🎯 Soru:
Bir listenin içinde kullanıcıdan alınan sayının olup olmadığını Linear Search kullanarak bulun.

✅ Çözüm:

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

nums = [10, 20, 30, 40, 50]
target = int(input("Aranacak sayıyı girin: "))

index = linear_search(nums, target)

if index != -1:
    print(f"{target} bulundu. İndeks: {index}")
else:
    print(f"{target} listede bulunamadı.")
🔎 2. Binary Search Egzersizi
🎯 Soru:
Sıralı bir liste içinde Binary Search kullanarak hedef sayıyı bulun.

✅ Çözüm:

def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1

arr = [5, 8, 12, 15, 22, 34, 60]
target = int(input("Aranacak sayıyı girin: "))

result = binary_search(arr, target)
if result != -1:
    print(f"{target} bulundu. İndeks: {result}")
else:
    print(f"{target} listede yok.")
🔃 3. Selection Sort Egzersizi
🎯 Soru:
Bir listeyi Selection Sort algoritmasıyla küçükten büyüğe sırala.

✅ Çözüm:

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]

liste = [64, 25, 12, 22, 11]
selection_sort(liste)
print("Sıralı liste:", liste)
🧩 4. Merge Sort Egzersizi
🎯 Soru:
Bir diziyi Merge Sort algoritması ile sırala.

✅ Çözüm:

def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])

    return merge(left_half, right_half)

def merge(left, right):
    sorted_list = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            sorted_list.append(left[i])
            i += 1
        else:
            sorted_list.append(right[j])
            j += 1

    sorted_list += left[i:]
    sorted_list += right[j:]

    return sorted_list

arr = [38, 27, 43, 3, 9, 82, 10]
sorted_arr = merge_sort(arr)
print("Sıralı liste:", sorted_arr)
⚡ 5. Quick Sort Egzersizi
🎯 Soru:
Quick Sort algoritması ile bir listeyi sıralayın.

✅ Çözüm:

def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[0]
    left = [x for x in arr[1:] if x < pivot]
    right = [x for x in arr[1:] if x >= pivot]

    return quick_sort(left) + [pivot] + quick_sort(right)

liste = [29, 10, 14, 37, 13]
print("Sıralı liste:", quick_sort(liste))
```
